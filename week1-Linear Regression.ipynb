```python
import pandas
from pandas import DataFrame
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data=pandas.read_csv('cost_revenue_dirty1.csv')
```

```python
data.describe()
```

```python
X=DataFrame(data, columns=['Production_Budget'])
y=DataFrame(data, columns=['Worldwide_Gross'])
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X,y, alpha=0.3)
plt.title('The genz vs hhh')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.xlabel('Production_Budget')

plt.ylabel('Worldwide_Gross')
plt.show()
```

```python
regression=LinearRegression()
regression.fit(X,y)
```

```python
regression.coef_
```

```python
regression.intercept_
```
